## Introduction

21st Century Americans are facing an increasingly politicized landscape, with norms and conventions being challenged at every turn. One of the most resoundingly partisan issues is the subject of gun ownership. American gun ownership is protected under the 2nd Amendment of the Constitution, but is with greater frequency being questioned in light of seemingly increasing violence. This is a contentious political issue, often with impassioned arguments on either side of the aisle. This report attempts to set aside partisan notions to as objectively as possible address trends based on accessible data. 

One source of gun data is the Federal Bureau of Investigation's National Instant Criminal Background Check System. In their own words:

> The National Instant Criminal Background Check System, or NICS, is all about saving lives and protecting people from harm—by not letting guns fall into the wrong hands. It also ensures the timely transfer of firearms to eligible gun buyers. 

>Mandated by the Brady Handgun Violence Prevention Act of 1993 and launched by the FBI on November 30, 1998, NICS is used by Federal Firearms Licensees (FFLs) to instantly determine whether a prospective buyer is eligible to buy firearms. Before ringing up the sale, cashiers call in a check to the FBI or to other designated agencies to ensure that each customer does not have a criminal record or isn’t otherwise ineligible to make a purchase. More than 230 million such checks have been made, leading to more than 1.3 million denials. [1]

Pairing this dataset with a dataset of census data will allow us to begin to formulate hypotheses about how a proxy for gun sales and demographics interact. This data can be useful to determine a baseline set of statistics about the basics of guns in America, such as whether gun registration is rising or declining, which states have the highest rate of gun registration, and, how gun registration is affected by levels of income/poverty. 

It should be stressed that none of these claims can be made with any amount of certainty and may only provide correlational and not causational trends. In order to make claims with a high degree of fidelity and vigor, hypothesis testing or regression analysis would have to be used.

### Data Cleaning

#### Gun Data

The first order of business will be to clean the Gun Data set. The dataset is very spotty, with many missing fields and empty values. In an attempt to preserve some integrity of the dataset, cleaning should be as minimal as possible. That said, some fields can confidently be cleaned. First is dropping the five values that are territories and not part of the 50 states (the Mariana Islands, Guam, Virgin Islands, Puerto Rico, and District of Columbia).

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.

#Drop the five values discussed above
gun_data = gun_data[gun_data.state != 'Guam']
gun_data = gun_data[gun_data.state != 'District of Columbia']
gun_data = gun_data[gun_data.state != 'Mariana Islands']
gun_data = gun_data[gun_data.state != 'Virgin Islands']
gun_data = gun_data[gun_data.state != 'Puerto Rico']

# Check values are dropped
gun_data['state'].unique()
```

#### Census Data

For the census data, we first want to remove the duplicated values.

```python
#dedupe data
censdata = censdata.drop_duplicates()
```

```python
#check censdata for removed duplicated values
sum(censdata.duplicated())
```

Removing the fact note column is next. Upon closer inspection, it seems like there are three rows that offer no new information - 'not available', 'not applicable' and 'suppressed.' These rows add no data points by design and should be removed. 

```python
#Drop rows with fact notes that do not add data points
censdata = censdata[censdata['Fact Note'] != 'Not available']
censdata = censdata[censdata['Fact Note'] != 'Suppressed; does not meet publication standards']
censdata = censdata[censdata['Fact Note'] != 'Suppressed to avoid disclosure of confidential information']
censdata = censdata[censdata['Fact Note'] != 'Not applicable']

#Check rows were removed
print(censdata['Fact Note'].unique())
```

```python
#Remove Fact Note column
censdata = censdata.drop('Fact Note', axis=1)
censdata.head()
```

## Exploratory Data Analysis

It is recommended that you be systematic with your approach. Look at one variable at a time, and then follow it up by looking at relationships between variables.

### What State has the highest gun registration? The lowest?

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
handgun_state = gun_data.groupby(['state']).sum()['handgun'].sort_values(ascending=False)
longgun_state = gun_data.groupby(['state']).sum()['long_gun'].sort_values(ascending=False)
othergun_state = gun_data.groupby(['state']).sum()['other'].sort_values(ascending=False)
multigun_state = gun_data.groupby(['state']).sum()['multiple'].sort_values(ascending=False)

plt.figure(1, figsize = (50,20))
plt.subplot(411)
ax = handgun_state.plot(kind='bar')
plt.title('Handguns: State vs. Count')
plt.xlabel('State')
plt.ylabel('Count');
```

```python
plt.figure(1, figsize = (50,20))
plt.subplot(412)
bx = longgun_state.plot(kind='bar')
plt.title('Long Guns: State vs. Count')
plt.xlabel('State')
plt.ylabel('Count');
```

```python
plt.figure(1, figsize = (50,20))
plt.subplot(412)
cx = othergun_state.plot(kind='bar')
plt.title('Other Gun Purchases: State vs. Count')
plt.xlabel('State')
plt.ylabel('Count');
```

```python
plt.figure(1, figsize = (50,20))
plt.subplot(412)
dx = multigun_state.plot(kind='bar')
plt.title('Multiple Gun Purchases: State vs. Count')
plt.xlabel('State')
plt.ylabel('Count');
```

It appears that Texas has the largest amount of handguns registered at roughly 600,000 guns, comes in second for long guns and other guns, and leads in multiple gun purchases (as would be expected by those metrics). By contrast, Hawaii seems to have the least registration in all categories accept 'other guns', of which it is the third least registration. This could be evident of lower sales, but it could also be representative of laxer legislation concernign registration. Thus these findings must not be taken as indicative of gun purchases or representative of the population's attitudes towards guns or gun legislation. 

### Has gun registration gone up or down with time? Are there milestones (spikes or dips in the time-series data)?

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
handgun_counts = gun_data.groupby(['month']).sum()['handgun']
longgun_counts = gun_data.groupby(['month']).sum()['long_gun']
othergun_counts = gun_data.groupby(['month']).sum()['other']
multigun_counts = gun_data.groupby(['month']).sum()['multiple']

plt.figure(1, figsize = (20,20))

plt.subplot(411)
handgunplt = handgun_counts.plot(kind='line') 
plt.title('Handgun Registration vs. Time')
plt.xlabel('Time')
plt.ylabel('Count');
```

```python
plt.figure(1, figsize = (20,20))

plt.subplot(412)
longgunplt = longgun_counts.plot(kind='line') 
plt.title('Long Gun Registration vs. Time')
plt.xlabel('Time')
plt.ylabel('Count');
```

```python
plt.figure(1, figsize = (20,20))

plt.subplot(413)
othergunplt = othergun_counts.plot(kind='line') 
plt.title('Other Gun Registration vs. Time')
plt.xlabel('Time')
plt.ylabel('Count');
```

```python
plt.figure(1, figsize = (20,20))

plt.subplot(414)
multigunplt = multigun_counts.plot(kind='line') 
plt.title('Multiple Gun Registration vs. Time')
plt.xlabel('Time')
plt.ylabel('Count');
```

Some interesting trends can be seen in the above graphs. Firstly, Handguns spike periodically but still show an upward trajectory in baseline, meaning there is a clear rise in handgun registration. Long Gun registration is similarly periodic, but does not have the upward baseline trend, indicating that long gun registration is somewhat stable. The other gun category only seems to have bee a relativley recent addition to the NICS background check, appearing for the first time in around 2009. This trend seems to be rising, but has the least values associated with it so can not be held to the same rigor as the insight provided from the other trends. Lastly, the multiple gun registration trends shows a relatively stable trend up until around 2007/2008 when it spikes to a high of 50,000 and has periodic but nonetheless arresting levels of registration. Prior to 2007, there appears to be no instance of multiple gun registrations upwards of 20,000 but post-2007 multiple gun registrations of 20,000 occur seemingly at least once a year. 

### Is gun registration effected by levels of income/poverty levels? 

```python
#make dataframe of relevant details
censincome = censdata.transpose()
medInc = censincome[47]
PerCapInc = censincome[48]
PovRates = censincome[49]

#Make readable
medInc = medInc.drop('Fact', axis=0).replace('[\$,]', '', regex=True).astype(float)
PerCapInc = PerCapInc.drop('Fact', axis=0).replace('[\$,]', '', regex=True).astype(float).sort_values(ascending=False)
PovRates = PovRates.drop('Fact', axis=0).replace('[\%,]', '', regex=True).astype(float).sort_values(ascending=False)
#The last 12 rows were still in decimal form, so I spot cleaned them here - not elegant, but it works
PovRates[-12:]=PovRates[-12:]*100
PovRates = PovRates.sort_values(ascending=False)
```

We must first bring in the census data. Due to the size of the census data, we will need to parse it down to what we need. For this investigation, I wished to look at rough measures of class on gun sales/background checks so I chose median income, per capita income and poverty. These are not perfect, or even good, measures of class but they offer enough to provide some insight into how access to capital might influence gun ownership.

```python
fig = plt.figure(figsize = (50,20)) # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}
handgun_alpha = gun_data.groupby(['state']).sum()['handgun']
#Combine data
combined = pd.concat([handgun_alpha, medInc], axis=1)
combined = combined.sort_values(47, ascending=False)
#Plot graphs
combined['handgun'].plot(kind='bar', color='red', ax=ax, width=bar_width, position=1)
combined[47].plot(kind='bar', color='blue', ax=ax2, width=bar_width, position=0)
#Set labels
ax.set_xlabel('States')
ax.set_ylabel('Handgun Counts')
ax2.set_ylabel('Median Income')
ax.set_title('Count of Handgun Background Checks vs. Median Incomce')
ax.legend()

fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (50,20)) # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}
handgun_alpha = gun_data.groupby(['state']).sum()['handgun']
#combine Data
combined = pd.concat([handgun_alpha, PerCapInc], axis=1)
combined = combined.sort_values(48, ascending=False)
#plot graphs
combined['handgun'].plot(kind='bar', color='red', ax=ax, width=bar_width, position=1)
combined[48].plot(kind='bar', color='blue', ax=ax2, width=bar_width, position=0)
#set labels
ax.set_xlabel('States')
ax.set_ylabel('Handgun Counts')
ax2.set_ylabel('Per Capita Income')
ax.set_title('Count of Handgun Background Checks vs. Per Capita Income')
ax.legend()

fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (50,20)) # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}
handgun_alpha = gun_data.groupby(['state']).sum()['handgun']
#combine data
combined = pd.concat([handgun_alpha, PovRates], axis=1)
combined = combined.sort_values(49, ascending=False)
#plot graph
combined['handgun'].plot(kind='bar', color='red', ax=ax, width=bar_width, position=1)
combined[49].plot(kind='bar', color='blue', ax=ax2, width=bar_width, position=0)
#set labels
ax.set_xlabel('States')
ax.set_ylabel('Handgun Counts')
ax2.set_ylabel('Poverty Rate')
ax.set_title('Count of Handgun Background Checks vs. Poverty Rate')
ax.legend()

fig.tight_layout()
plt.show()
```
