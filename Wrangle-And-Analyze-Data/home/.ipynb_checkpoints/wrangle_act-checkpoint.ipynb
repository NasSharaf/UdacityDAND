And, the most difficult part (in my opinion), making a file from the Twitter API. First we access the API. In the turned in documents, this will be blanked out.

```python
# gathering twitter data from API
import tweepy

consumer_key = ####
consumer_secret = ####
access_token = ####
access_secret = ####

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser(), wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

With the API authenticated, we can begin the process to query tweets.

```python
# Make twitter api text file
df_list = []
tweet_errors = []

for tweetid in tweetids:
    # tweetjson = tweet._json
    try:
        #make the tweet text file
        tweet = api.get_status(tweetid)
        with open('tweet_json.txt', 'w') as outfile:  
            json.dump(tweet, outfile)
        #read in from text file and make df
        with open('tweet_json.txt') as json_file:
            data = json.load(json_file)
            tweetID = data['id']
            fav_count = data['favorite_count']
            retweet_count = data['retweet_count']
            #followers = data['followers_count']
            # sensitive = data['possibly_sensitive']
            df_list.append({'Tweet_ID': tweetID,
                            'fav_count': fav_count,
                            'retweet_count': retweet_count})
    #Error handling attempt
    except tweepy.TweepError as e:
        tweet_errors.append(tweetid)  
```

With our data collected, our task is to convert the data set into a dataframe and take a look.

NOTE: After attaining the tweepy values, I used the tweet_json.txt file to save time and not have to run half an hour of code each time.

## Cleaning Data

Our first step is to make copies of all the dataframes.

```python
df_clean = df.copy()
```

```python
df_image_clean = df_image.copy()
```

```python
df_api_clean = df_api.copy()
```

Generally, we would like to address the NaNs first, but since several of those are related to the retweets, I'd rather get rid of the retweets first and then address missing values. Also, I intend to merge the dataframes onto the archived twitter data, which will be easier if the retweets are already edited out of the dataset.

#### Define

Remove retweeted entries

#### Code

```python
df_clean = df_clean[df_clean.retweeted_status_id.isnull()]
```

### Tidiness

##### Drop NaN Columns

#### Define
Drop the in_reply_to columns and the Retweeted columns

#### Code

```python
df_clean = df_clean.drop('in_reply_to_status_id', axis=1)
df_clean = df_clean.drop('in_reply_to_user_id', axis=1)
df_clean = df_clean.drop('retweeted_status_id', axis=1)
df_clean = df_clean.drop('retweeted_status_user_id', axis=1)
df_clean = df_clean.drop('retweeted_status_timestamp', axis=1)
```

#### Test

```python
df_clean.info()
```

##### Merge Three dataframes into one

#### Define

Merge df_clean, df_image_clean, and df_api_clean. All three datasets are robust enough to be merged now. 

#### Code

```python
df_master1 = pd.merge(df_clean, df_image_clean, on=['tweet_id','tweet_id'], how='inner')
df_api_clean = df_api_clean.rename(index=str, columns={"Tweet_ID": "tweet_id"})
df_master = pd.merge(df_master1, df_api_clean, on=['tweet_id','tweet_id'], how='left')
```
