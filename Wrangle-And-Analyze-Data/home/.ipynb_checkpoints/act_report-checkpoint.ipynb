### WeRateDogs Report

It appears that the popular Twitter account WeRateDogs has a steady fanbase. While the account doesn't quite live up to it's initial fanfare and the enthusiasm has died down, the account is still going strong with occasional bursts of high activity. The popular account humorously rates pictures of dogs and rates them very highly, frequently giving ratings that exceed 10 out of 10. Why do the dogs deserve such high ratings? Because they're good dogs of course.



We can see in the favorites chart that while the highs will still rise up to the 80,000 mark, it never reaches the past the 100,000 favorites like it used to do. There is still a clear fanbase, but it is no longer as supportive as it was in the heyday of WeRateDogs. 



There is a similar story with retweets, although not as prominent. In particular, there is a spike at tweet index 794, which was a particularly poignant tweet about one of the dogs who responded in 9/11 who had passed away. WeRateDogs continues to put forward a genuinely sincere love of animals through their humor context.
